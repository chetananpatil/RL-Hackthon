```python
import gym
import numpy as np

# Create the FrozenLake environment
env = gym.make('FrozenLake-v1')

# Initialize Q-table with zeros
action_space_size = env.action_space.n
state_space_size = env.observation_space.n
q_table = np.zeros((state_space_size, action_space_size))

# Hyperparameters
num_episodes = 10000
max_steps_per_episode = 100
learning_rate = 0.1
discount_rate = 0.99
exploration_rate = 1
max_exploration_rate = 1
min_exploration_rate = 0.01
exploration_decay_rate = 0.001

# Track rewards for each episode
total_rewards = []

# Q-learning algorithm
for episode in range(num_episodes):
    state = env.reset()
    done = False
    rewards_current_episode = 0
    episode_steps = []

    for step in range(max_steps_per_episode):
        # Exploration-exploitation trade-off
        exploration_rate_threshold = np.random.uniform(0, 1)
        if exploration_rate_threshold > exploration_rate:
            action = np.argmax(q_table[state, :])
        else:
            action = env.action_space.sample()

        # Take action and observe new state & reward
        new_state, reward, done, _ = env.step(action)

        # Update Q-table for Q(s, a)
        q_table[state, action] = q_table[state, action] * (1 - learning_rate) + \
            learning_rate * (reward + discount_rate * np.max(q_table[new_state, :]))

        episode_steps.append(action)
        state = new_state
        rewards_current_episode += reward

        if done:
            print(f"Episode {episode + 1} done.")
            if reward == 1:
                print("Goal reached!")
            else:
                print("Agent fell into a hole.")
            break

    total_rewards.append(rewards_current_episode)
    print(f"Total rewards for Episode {episode + 1}: {rewards_current_episode}")
    print(f"Path taken in Episode {episode + 1}: {episode_steps}")

    # Exploration rate decay
    exploration_rate = min_exploration_rate + \
        (max_exploration_rate - min_exploration_rate) * np.exp(-exploration_decay_rate * episode)

# Calculate average total reward
average_total_reward = np.mean(total_rewards)
print(f"\nAverage total reward: {average_total_reward}")

env.close()
```
